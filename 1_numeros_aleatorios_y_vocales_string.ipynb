# Case 1

```python
#Importamos la libreria random para generar un aleatorio de numeros
import random

# en input_array guardaremos los numeros generados
input_array = []
#con un ciclo for generaremos 10 numeros aleatorios del 1 al 10
for i in range(10):
  input_array.append(random.randint(1,10))

print('array de numeros ==> ',input_array)
print('largo del array ==> ', len(input_array))

#En el diccionario ocurrencias guardaremos el numero de veces que esta cada numero generado y guardado en input_array
ocurrencias = {}
for val in input_array:
  if val in ocurrencias:
    ocurrencias[val] += 1
  else:
    ocurrencias[val] = 1


print('numeros y cantidad de ocurrencias ==>', ocurrencias)

# Buscamos el número con mayor cantidad de ocurrencias
max_ocurrencias = max(ocurrencias.values())
print(max_ocurrencias)

#Mostramos todos los números que tienen la cantidad máxima de ocurrencias
numeros_max_ocurrencias = []
for numero, cant_ocurrencias in ocurrencias.items():
  if cant_ocurrencias == max_ocurrencias:
    numeros_max_ocurrencias.append(numero)

# Mostramos el resultado
print('Numero(s) con mas ocurrencias ==> ',numeros_max_ocurrencias)
```

```python
#Importamos la libreria random para aleatorizar la cadena de texto, y el modulo string para generar letras, 
import random, string

# En la variable texto almacenaremos el texto
texto= ""
# En el ciclo for generaremos la cadena de texto con 10 letras en minuscula
for val in range(10):
  texto += random.choice(string.ascii_lowercase)

print(texto)
# Definimos una función para verificar si un caracter es una vocal
def es_vocal(caracter):
    vocales = ["a", "e", "i", "o", "u"]
    return caracter in vocales

# Contamos el número de vocales en el string
cantidad_vocales = 0
for caracter in texto:
    if es_vocal(caracter):
        cantidad_vocales += 1

# Mostramos el resultado
print("Cantidad de vocales en el string:", cantidad_vocales)
```
